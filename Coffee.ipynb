Data Visualization

```python
top_import = coffee_import[['Country','Total_import']].sort_values(by='Total_import',ascending = False).head(10)
top_export = coffee_export[['Country','Total_export']].sort_values(by='Total_export',ascending = False).head(10)
top_consu = coffee_domestic_consumption[['Country','Total_domestic_consumption']].sort_values(by='Total_domestic_consumption',ascending = False).head(10)
top_imp_consu = coffee_importers_consumption[['Country','Total_import_consumption']].sort_values(by='Total_import_consumption',ascending = False).head(10)
top_prod = coffee_production[['Country','Total_production']].sort_values(by='Total_production',ascending = False).head(10)
top_inv = coffee_green_coffee_inventories[['Country','Total_inventorie']].sort_values(by='Total_inventorie',ascending = False).head(10)

fig = make_subplots(rows=1,cols=5,shared_yaxes=False,subplot_titles=('Top Importers','Top Exporters','Top Consumption','Top Production','Top Import Consumption'))
fig.add_bar(x=top_import['Country'],y=top_import['Total_import'],row=1,col=1,showlegend=False)
fig.add_bar(x=top_export['Country'],y=top_export['Total_export'],row=1,col=2,showlegend=False)
fig.add_bar(x=top_consu['Country'],y=top_consu['Total_domestic_consumption'],row=1,col=3,showlegend=False)
fig.add_bar(x=top_prod['Country'],y=top_prod['Total_production'],row=1,col=4,showlegend=False)
fig.add_bar(x=top_imp_consu['Country'],y=top_imp_consu['Total_import_consumption'],row=1,col=5,showlegend=False)
```

Trend of Top Countries

```python
def transpose(df,sort_val,additional_col=None):
  df_top = df.sort_values(by=sort_val,ascending = False).head(n=10)
  df = df_top.T
  df.drop(['Country',sort_val],inplace=True,axis=0)
  if additional_col != None:
    df.drop(additional_col)
  df.columns = df_top['Country']
  return df
```

```python
import_df = transpose(coffee_import,'Total_import')
print(import_df.head())
export_df = transpose(coffee_export,'Total_export')
cons_df = transpose(coffee_domestic_consumption,'Total_domestic_consumption')
pro_df = transpose(coffee_production,'Total_production')
```

```python
import_df
fig = px.line(import_df,x=import_df.index,y=import_df.columns,title='Top Importer')
fig.show()
fig = px.line(export_df,x=export_df.index,y=export_df.columns,title='Top Exporter')
fig.show()
fig = px.line(cons_df,x=cons_df.index,y=cons_df.columns,title='Top Consumption')
fig.show()
fig = px.line(pro_df,x=pro_df.index,y=pro_df.columns,title='Top Production')
fig.show()
```
